```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
# Load the Titanic dataset from Seaborn
titanic_df = sns.load_dataset('titanic')
```

```python
# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Count the number of man, woman and child
counts = titanic_df['who'].value_counts()
# Plot the distribution of man, woman and child using a pie chart
sns.set_style("darkgrid")
sns.set_palette("colorblind")
sns.set_context("notebook", font_scale=1.2)
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
# Countplot
sns.catplot(x ="who", hue ="survived",kind ="count", data = titanic_df)
```

```python
# Countplot for passenger gender
sns.countplot(data=titanic_df, x='sex',hue='survived')
```

```python
#Plot a histogram of the ticket prices
sns.histplot(data=titanic_df, x='fare', bins=30,kde=True)
plt.title('Distribution of Ticket Fares')
plt.show()
```

```python
sns.jointplot(data=titanic_df,y='fare',x='survived',hue='sex')
```

```python
#Heatmap: A heatmap is useful to visualize the correlation between variables in a dataset.
# Let's create a correlation matrix and plot a heatmap to explore the relationships between different numerical variables in the Titanic dataset

# Calculate the correlation matrix
correlation_matrix = titanic_df.corr()
# Plot a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
```
